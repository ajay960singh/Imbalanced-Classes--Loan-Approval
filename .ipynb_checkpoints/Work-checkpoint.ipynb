# Import the relevant packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
```

# Get Data and Perform Pre-processing

```python
# Create a dataframe fro data available
df=pd.read_csv('train-2.csv')
```

```python
# Check if data is missing
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis') # yellow lines show the missing data points
```

```python
# Check if classes are imbalanced
sns.countplot(x='Approved',data=df)
```

```python
# Imbalanced classes
df['Approved'].value_counts()
```

```python
# Find the count of a number of different missing data points
sns.countplot(df.isnull().sum(axis=1))
```

```python
# Create a new df with less than 7 missing values
df1=df[(df.isnull().sum(axis=1)<7)]
```

```python
# Get the number of missing values
df1.isnull().sum(axis=0)
```

```python
sum(df1[df1['Interest_Rate'].isnull()]['Approved']==1)
```

```python
sum(df1[df1['EMI'].isnull()]['Approved']==1)
```

##### Can't drop Interest_Rate & EMI as the null values contain abt 50% of approved loans (minority class)

```python
# Create a function to impute the missing values
def impute(col):
    mean_temp_no=df1[df1['Approved']==0][col].mean()
    df1.loc[df1['Approved']==0,col]=df1[df1['Approved']==0][col].fillna(mean_temp_no)
    mean_temp_yes=df1[df1['Approved']==1][col].mean()
    df1.loc[df1['Approved']==1,col]=df1[df1['Approved']==1][col].fillna(mean_temp_yes)
```

```python
impute('Loan_Amount')
impute('Interest_Rate')
impute('EMI')
impute('Existing_EMI')
```

```python
# Identufy the categorical features with mssing values
cat_feats=['Employer_Code','Employer_Category1','Employer_Category2','Customer_Existing_Primary_Bank_Code','Primary_Bank_Type','City_Code','City_Category','Loan_Period']
```

```python
for cat in cat_feats:
    print(df1[cat].value_counts())
```

```python
# Impute the most frequent classes to the missing data points
df1['Employer_Category1'].fillna('A',inplace=True)
df1['Employer_Category2'].fillna('4.0',inplace=True)
df1['Customer_Existing_Primary_Bank_Code'].fillna('B001',inplace=True)
df1['Primary_Bank_Type'].fillna('P',inplace=True)
df1['City_Category'].fillna('A',inplace=True)
df1['Loan_Period'].fillna('5.0',inplace=True)
df1['City_Code'].fillna('C10001',inplace=True)
```

```python
# Check for missing values finally
sns.heatmap(df1.isnull(), yticklabels=False, cbar=False, cmap='viridis') 
```

## Adding new features

```python
# Add a binary feature for 0s in Existing_EMI
df1['IsZero']=0
df1.loc[df1['Existing_EMI']==0,'IsZero']=1
```

```python
# Split the date column, first change them to string types
df1['DOB'] = df1['DOB'].astype(str)
df1['Lead_Creation_Date'] = df1['Lead_Creation_Date'].astype(str)

df1['dob_yr']=df1['DOB'].str[6:8]
df1['lead_yr']=df1['Lead_Creation_Date'].str[6:8]
```

```python
#Add a column, which has the difference between the years as entries
df1['dob_yr'] = df1['dob_yr'].astype(int)
df1['Diff']=(2016-(1900+df1['dob_yr']))
```

```python
# Drop the other columns
df1.drop(['DOB','Lead_Creation_Date','lead_yr','ID','dob_yr'],axis=1,inplace=True)
```

```python
num=['Monthly_Income','Existing_EMI','Loan_Amount','Approved','Interest_Rate','EMI']
# Get the categorical features
cate=[]
for i in df1.columns:
    if i not in num:
        cate.append(i)
print(cate)
```

```python
# Label Encoding of categorical features
for c in cate:
    df1[c]=pd.factorize(df1[c])[0]
```

```python
df1.T
```

# Train-test split 

```python
# Divide into independent and target features
X=df1.drop('Approved',axis=1)
y=df1['Approved']
```

```python
# Split into training and test using stratify
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)
```

# Up-sampling of minority class 

```python
# Combine the training set
df_temp = pd.concat([X_train, y_train],axis=1)
```

```python
# Separate majority and minority classes
df_majority = df_temp[df_temp.Approved==0]
df_minority = df_temp[df_temp.Approved==1]
```

```python
df_majority.shape
```

```python
# Upsample minority class
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=45122,    # to match majority class
                                 random_state=123) # reproducible results
```

```python
# Combine majority class with upsampled minority class
df_final = pd.concat([df_majority, df_minority_upsampled])
```

```python
# Check the count of classes
sns.countplot(x='Approved',data=df_final)
```

```python
# Split the training set
X_train=df_final.drop('Approved',axis=1)
y_train=df_final['Approved']
```

# Random Forest

```python
# Fit the training set to Random Forest model
rf=RandomForestClassifier(n_estimators=200,random_state=0,max_features=5).fit(X_train,y_train)
```

```python
# Predict
y_pred=rf.predict(X_test)
```

```python
# Get the F1 scores for the imbalanced classes
print(classification_report(y_pred,y_test))
```

#### Since our focus is to find the people, who will have a higher probability of getting their loans approved, Recall is more important than Precision So, even though, we might be taking 47% wrong people into consideration, we are getting 95% of the right people
